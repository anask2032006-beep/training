```python
s = input()

l = ""

for i in range(len(s)):
    cur = ""

    for j in range(i, len(s)):
        if s[j] not in cur:
            cur += s[j]
        else:
            break

    if len(cur) > len(l):
        l = cur

print(len(l))
```

```python
#polymorphism(overloading)
class multiply:
  def product(self, a=None, b=None, c=None):
    if a and b and c:
      print("product=",a*b*c)
    elif a and b:
      print("product=",a*b)
    else :
      print("invalid input")
obj = multiply()
obj.product(2,3)
obj.product(2,3,4)
```
